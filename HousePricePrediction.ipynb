**Applying normalization**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
new_data = scaler.fit_transform(df)
```

```python
normalized_df = pd.DataFrame(new_data,columns=df.columns)
```

```python
normalized_df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1 = normalized_df[['NOX','RM','DIS','PTRATIO','LSTAT' ]]
```

```python
y1 = normalized_df['MEDV']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x1, y1, test_size =0.2,random_state = 5 )
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```

```python
df
```

```python
from sklearn.metrics import r2_score

answer = r2_score(y_test,y_pred)
```
